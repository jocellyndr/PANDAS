# **Pandas 33 - Nilai Kumulatif pada Pandas Data Frame**

### Import Modules

```python
import pandas as pd

print(pd.__version__)
```

### Persiapan Data Frame

```python
d = {'pemain':['Budi', 'Joni', 'Iwan', 'Budi', 'Budi', 'Iwan', 'Asep', 'Joni'], 
     'goal':[2, 1, 3, 1, 1, 2, 2, 3]}
df = pd.DataFrame(d)
df
```

### **Menampilkan nilai kumulatif (cummulative)**
- Fungsi cumsum() untuk menampilkan data kumulatif
- Nilai kumulatif = nilai saat ini ditambahkan jumlah nilai sebelumnya

```python
# Menggunakan metode cumsum() pada kolom 'goal' dan mengonversi hasilnya ke DataFrame
df['goal'].cumsum().to_frame()
```

```python
# menambah kolom baru
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
```

```python
# menambah kolom baru lagi
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

```python
# menambah kolom baru bernama 'cummax' yang akan mendata nilai tertinggi sampai titik tertentu
df['cummax'] = df['goal'].cummax()
df
```

```python
# menambah kolom baru bernama 'cummin' untuk mendata nilai terendah sampai titik tertentu
df['cummin'] = df['goal'].cummin()
df
```

```python
# menambah kolom baru bernama 'cumprod' (cummulative product), yaitu hasil total perkalian/ perkalian kumulatif
df['cumprod'] = df['goal'].cumprod()
df
```
